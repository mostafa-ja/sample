```python
import torch
import torch.nn as nn
```

# Binary class

binary.jpg



```python
class NeuralNet1(nn.Module):
  def __init__(self,input_size,hidden_size):
    super().__init__()
    self.linear1 = nn.Linear(input_size,hidden_size)
    self.relu = nn.ReLU()
    self.linear2 = nn.Linear(hidden_size,1)
                                            # Option B , define sigmoid here , self.sigmoid = nn.Sigmoid()
                                        #ATTENTION : in __init__ part we use nn. .. module but for forward we use torch. ... like nn.Sigmoid and torch.sigmoid
  def forward(self,x):
    out = self.linear1(x)
    out = self.relu(out)
    out = self.linear2(out)
    # sigmoid at the end
    y_pred = torch.sigmoid(out)                   # y_pred = self.sigmoid(out)
    return y_pred 
```

```python
model = NeuralNet1(input_size=28*28 , hidden_size=5)
criterion = nn.BCELoss()
```

# Multi class
multiclass.jpg

```python
class NeuralNet2(nn.Module):

  def __init__(self,input_size,hidden_size,num_classes):
    super().__init__()
    self.linear1 = nn.Linear(input_size,hidden_size)
    self.relu = nn.ReLU()
    self.linear2 = nn.Linear(hidden_size,num_classes)

  def forward(self,x):
    out = self.linear1(x)
    out = self.relu(out)
    out = self.linear2(out)
    # ATTENTION : no softmax at the end , because of nn.CrossEntropyLoss() which needs raw scors(logits) without softmax 
    return out
  
```

```python
model = NeuralNet2(input_size=28*28,hidden_size=5,num_classes=3)
criterion = nn.CrossEntropyLoss()
```
