This are the raw attributes that are used for calculating how much damage an attack will do in the games. This dataset is about the pokemon games (NOT pokemon cards or Pokemon Go).

The data as described by Myles O'Neill is:

Name: Name of each pokemon

Type 1: Each pokemon has a type, this determines weakness/resistance to attacks

Type 2: Some pokemon are dual type and have 2

Total: sum of all stats that come after this, a general guide to how strong a pokemon is

HP: hit points, or health, defines how much damage a pokemon can withstand before fainting

Attack: the base modifier for normal attacks (eg. Scratch, Punch)

Defense: the base damage resistance against normal attacks

SP Atk: special attack, the base modifier for special 
attacks (e.g. fire blast, bubble beam)

SP Def: the base damage resistance against special attacks

Speed: determines which pokemon attacks first each round

# **Aggregate Statistics using Groupby (Sum, Mean, Counting)**

```python
#reset data to default
df = pd.read_csv('Pokemon.csv')
df
```

```python
df.groupby(['Type 1']).mean()   #groupby: group data based on (here , type 1) and then use information like mean of each group
```

```python
df.groupby(['Type 1']).mean().sort_values('Attack', ascending=False)
```

```python
df.groupby(['Type 1']).count()
```

```python
# better way to count
# first create a new column for count then use just that column
df['count']=1
df.groupby(['Type 1']).count()['count']  # ['count'] means we want just the 'count' column
```

```python
# how many of type one characters are related to one specific type 2 
df.groupby(['Type 1','Type 2']).count()['count']
```

# **Working with large amounts of data (setting chunksize)**

```python
for df in pd.read_csv('Pokemon.csv',chunksize=50):
  print(df.groupby(['Type 1']).mean())

```
